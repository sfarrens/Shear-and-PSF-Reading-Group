````python
# Import dependencies
import numpy as np
from plot import *
from data import *
from moments import *
import galsim
from IPython.display import Markdown as md
from interact import get_url, binder_badge

# Provide binder badge
url = get_url('moments-weighted')
md('<a href="{}" target="_blank"><img src="{}"></a>'.format(get_url('moments'), binder_badge))
````

````python
def show_ellipse(a, b, theta, sersic=1, sigma=0.0):
    
    data = (galsim.Sersic(n=sersic, scale_radius=1).shear(galsim.Shear(q=b/a, 
            beta=-theta*galsim.degrees)).drawImage(nx=50, ny=50).array)
    data += np.random.randn(*data.shape) * sigma
    centroid = get_centroid(data)
    chi = get_ellipticity(data)
    epsilon = get_ellipticity(data, method='epsilon')
    abt = get_abt(data)
    
    show_image(data, centroid=centroid, chi=chi, epsilon=epsilon, abt=abt)
````

````python
show_ellipse(1, 0.5, 45)
````
